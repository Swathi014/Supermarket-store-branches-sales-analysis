```python
Area_bins = [740, 1320, 1654, 2230]
Area_labels = ['Low', 'Medium', 'High']
df['Store_Area_C'] = pd.cut(df['Store_Area'], bins=Area_bins, labels=Area_labels, right=False)


plt.bar(df['Store_Area_C'],df['Daily_Customer_Count'] , width=0.15, color='green')
plt.xlabel('Store Size')
plt.ylabel('Daily Customer Count')
plt.title('Daily Customer Count')

plt.legend()
```

```python
plt.hist(df['Store_Area'], bins=30, color='green', edgecolor='black')
plt.xlabel('Store Area')
plt.ylabel('Frequency')
plt.title('Store Area')

plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
# Scatter plot for Store Area vs. Items Available
plt.figure(figsize=(8, 6))
plt.scatter(df['Store_Area'], df['Items_Available'], color='blue', alpha=0.5)
plt.title('Store Area vs. Items Available')
plt.xlabel('Store Area (sq. units)')
plt.ylabel('Items Available')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Store Area vs. Daily Customer Count
plt.figure(figsize=(8, 6))
plt.scatter(df['Store_Area'], df['Daily_Customer_Count'], color='red', alpha=0.5)
plt.title('Store Area vs. Daily Customer Count')
plt.xlabel('Store Area (sq. units)')
plt.ylabel('Daily Customer Count')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Store Area vs. Store Sales
plt.figure(figsize=(8, 6))
plt.scatter(df['Store_Area'], df['Store_Sales'], color='green', alpha=0.5)
plt.title('Store Area vs. Store Sales')
plt.xlabel('Store Area (sq. units)')
plt.ylabel('Store Sales')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Items Available vs. Daily Customer Count
plt.figure(figsize=(8, 6))
plt.scatter(df['Items_Available'], df['Daily_Customer_Count'], color='purple', alpha=0.5)
plt.title('Items Available vs. Daily Customer Count')
plt.xlabel('Items Available')
plt.ylabel('Daily Customer Count')
plt.grid(True)
plt.show()
```

```python
# Scatter plot for Items Available vs. Store Sales
plt.figure(figsize=(8, 6))
plt.scatter(df['Items_Available'], df['Store_Sales'], color='orange', alpha=0.5)
plt.title('Items Available vs. Store Sales')
plt.xlabel('Items Available')
plt.ylabel('Store Sales')
plt.grid(True)
plt.show()
```

The scatter plots show these relationships: bigger stores usually have more items available, but store size doesn't always mean more customers; bigger stores often make more sales, and stores with more items available tend to make more sales too; stores with more customers typically make more sales.


# Conclusion


1. Size doesn't equal more customers: Just because a store is bigger doesn't mean it gets more visitors.

2. Profit isn't guaranteed: Larger stores might make more money overall, but they might not make more money per customer. Sometimes, smaller stores are better at making each customer spend more.


3. So, store size, items available, and customer count all affect how much a store sells.


1. Efficiency matters: It's not just about how much space a store has. Retailers need to think about how efficiently they use that space to make sure they're making the most money they can.
They should focus on making stores attractive and easy to shop in to keep customers coming back.
